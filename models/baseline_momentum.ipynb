```python
import pandas as pd
import numpy as np
df = pd.read_csv('..\\baseline_data\\stock_price_monthly_2003_2024.csv')
```

```python
# Drop rows with NaN past returns (first 12 months per stock)
df = df.dropna(subset=["past_return"])

# only get from jan 2004
start_date = '2005-01'
year = start_date.split('-')[0]
month = start_date.split('-')[1]

df1 = df[df['date'] == start_date]
df1
```

```python
# Rank stocks based on past return (higher is better)
df1["rank"] = df1.groupby("date")["past_return"].rank(method="first", ascending=False)
df1 = df1.sort_values(by=["date", "rank"])
```

```python
# Define the winner, loser (top 10% ,bottom 10%), and neutral portfolios
df1["portfolio"] = df1.groupby("date")["rank"].transform(
    lambda x: np.where(x <= np.percentile(x, 10), "W", 
                       np.where(x >= np.percentile(x, 90), "L", "N"))
)
```

```python
# save
# df1.to_csv(f'..\\baseline_data\\stock_price_monthly_{year}_{month}.csv', index=False)
```

```python
buy = df1[df1["portfolio"] == "W"]
sell = df1[df1["portfolio"] == "L"]

money = 0

spend = buy["PRC"].sum() - sell["PRC"].sum()
money -= spend

money
```

```python
# 6 months later
end_date_pd = pd.to_datetime(start_date) + pd.DateOffset(months=6)
end_date = end_date_pd.strftime('%Y-%m')

print(end_date)
df2 = df[df['date'] == end_date]
df2
```

```python
# save df2
year = end_date.split('-')[0]
month = end_date.split('-')[1]
# df2.to_csv(f'..\\baseline_data\\stock_price_monthly_{year}_{month}.csv', index=False)
```

```python
# Rebalance the portfolio
# get all stocks price that were bought and sell earlier
rebalance_buy = df2[df2["NCUSIP"].isin(buy["NCUSIP"])]
rebalance_sell = df2[df2["NCUSIP"].isin(sell["NCUSIP"])]
spend = -rebalance_buy["PRC"].sum() + rebalance_sell["PRC"].sum()
money -= spend

spend
```
